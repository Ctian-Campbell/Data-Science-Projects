```python
og_crime_data = pd.read_csv("C:/Users/Owner/Documents/Important/Bellevue/5 - Data Preparation/Project/crime_and_incarceration_by_state.csv")
og_crime_data
```

##### 1.

```python
# In this step I will be deleting the rape_revised column.
# The database contains data from 2001 to 2016 however the rape_revised column has null values till 2012.
# As such, I find that column to be incomplete. Which is why I choose to delete it.
```

```python
step1_crime_data = og_crime_data.drop('rape_revised', axis=1)
step1_crime_data
```

##### 2.

```python
# The rows labelled Federal contain only 4 columns of data.
# The rest of the row contains null values. 
# For the purposes of this project, the "Federal" rows are not needed. Which is why I'm deleting them.
```

```python
step2_crime_data = step1_crime_data[step1_crime_data['jurisdiction'] != 'FEDERAL']
step2_crime_data
```

##### 3.

```python
# I have decided to convert the values in the include_jails column to integers. 
# False = 0 and True = 1
```

```python
step3_crime_data = step2_crime_data.copy()  
step3_crime_data['includes_jails'] = step3_crime_data['includes_jails'].astype(int)
step3_crime_data
```

##### 4.

```python
# I have decided to convert the values in the crime_reporting_change column to integers. 
# False = 0 and True = 1
```

```python
step4_crime_data = step3_crime_data.copy()  # Create a copy of the original DataFrame
step4_crime_data['crime_reporting_change'] = step4_crime_data['crime_reporting_change'].map({False: 0, True: 1})
step4_crime_data
```

##### 5.

```python
# I have decided to convert the values in the crime_estimated column to integers. 
# False = 0 and True = 1
```

```python
final_crime_data = step4_crime_data.copy()  
final_crime_data['crimes_estimated'] = step3_crime_data['crimes_estimated'].map({False: 0, True: 1})
final_crime_data
```
