##### 2.

```python
# Although, in the df above the states look like they're in alphabetical order. They actually aren't.
# This code chunk orders the State column in ascending alphabetical order.
fbi_step2_data = fbi_step1_data.sort_values(by='State')
fbi_step2_data.head(24)
```

##### 3.

```python
# In this third step I'm going to change the 2nd column header and make it look like the other columns.
fbi_step3_data = fbi_step2_data.rename(columns={'data_year': 'Year'})
fbi_step3_data.head(24)
```

##### 4.

```python
# In this fourth step, I'll be sorting the year column into ascending order so that it matches how the dtaframes from
# the other sources are sorted.
fbi_step4_data = fbi_step3_data.sort_values(by=['Year', 'State']).reset_index(drop=True)
fbi_step4_data.head(24)
```

##### 5.

```python
# This fourth step is about finding and deleting nan values. 
# The code below finds nan values.
rows_with_nan = fbi_step4_data[fbi_step4_data.isna().any(axis=1)]
print(f"Rows with NaN values:{rows_with_nan}")
```

```python
# The code below deletes the row with nan values.
index_to_delete = fbi_step4_data[
    (fbi_step4_data['State'] == 'Florida') & (fbi_step4_data['Year'] == 2016)].index
fbi_final_data = fbi_step4_data.drop(index_to_delete).reset_index(drop=True)

fbi_final_data.head(24)
```
